```python
#Question 3

#defining function 
def Common():
    #asking user for 2 strings and combinding them into 1 string
    String1 = input("Enter your first string:")
    String2= input("Enter your second string:")
    BothStr= String1 + " " + String2
    StrList = BothStr.split()
    SolutionStr = {}
    x=0
    
    #counting the common words and putting them in a dictionary
    while x<len(StrList):
        y=0
        i=0
        while i <len(StrList):
            if StrList[i] == StrList[x]:
                y=y+1
            i=i+1
        if y>=2 and StrList[x] not in SolutionStr:
            SolutionStr[StrList[x]]=y
        x=x+1
    #outputing common words and count
    print("Common words and their counts:")
    print(SolutionStr)
Common()   
```
